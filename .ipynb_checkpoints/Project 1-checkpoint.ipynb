```python
 # jupyter # Import required libraries
from flask import Flask, request
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import os
from werkzeug.serving import run_simple

# Create a Flask app
app = Flask(_name_)

# Load the pre-trained model (.h5) saved locally
model_path = r"C:\Users\Mayuri\Desktop\Medical plant identification\medicinal_plants_model.h5"  # Update to your model path

# Load the model
try:
    model = tf.keras.models.load_model(model_path)
    print(f"Model loaded successfully from {model_path}")
except Exception as e:
    print(f"Error loading model: {e}")

# Define class labels with actual species names
class_labels = [
    'Amaranthus Green',
    'Amaranthus Red',
    'Asthma Plant.zip',
    'Avaram.zip',
    'Balloon vine',
    'Balloon vine.zip',
    'Bellyache bush (Green).zip',
    'Benghal dayflower.zip',
    'Betel Leaves',
    'Big Caltrops.zip',
    'Black Night Shade',
    'Black-Honey Shrub.zip',
    'Bristly Wild Grape.zip',
    'Butterfly Pea.zip',
    'Cape Gooseberry.zip',
    'Celery',
    'Chinese Spinach',
    'Common Wireweed.zip',
    'Coriander Leaves',
    'Country Mallow.zip',
    'Crown flower.zip',
    'Curry Leaf',
    'Dwarf Copperleaf (Green)',
    'Dwarf copperleaf (Red)',
    'False Amarnath',
    'Fenugreek Leaves',
    'Giant Pigweed',
    'Gongura',
    'Green Chireta.zip',
    'Holy Basil.zip',
    'Indian CopperLeaf.zip',
    'Indian Jujube.zip',
    'Indian Sarsaparilla.zip',
    'Indian Stinging Nettle.zip',
    'Indian Thornapple.zip',
    'Indian pennywort',
    'Indian wormwood.zip',
    'Ivy Gourd.zip',
    'Kokilaksha.zip',
    'Lagos Spinach',
    'Lambs Quarters',
    'Land Caltrops (Bindii).zip',
    'Lettuce Tree',
    'Madagascar Periwinkle.zip',
    'Madras Pea Pumpkin.zip',
    'Malabar Catmint.zip',
    'Malabar Spinach (Green)',
    'Mexican Mint.zip',
    'Mexican Prickly Poppy.zip',
    'Mint Leaves',
    'Mountain Knotgrass.zip',
    'Mustard',
    'Nalta Jute.zip',
    'Night blooming Cereus.zip',
    'Palak',
    'Panicled Foldwing.zip',
    'Prickly Chaff Flower.zip',
    'Punarnava.zip',
    'Purple Fruited Pea Eggplant.zip',
    'Purple Tephrosia.zip',
    'Rosary Pea.zip',
    'Shaggy button weed.zip',
    'Siru Keerai',
    'coatbuttons.zip',
    'heart-leaved moonseed.zip'
]

# Ensure uploads directory exists
if not os.path.exists('uploads'):
    os.makedirs('uploads')

# Define the home route to render the upload page
@app.route('/')
def home():
    return '''
    <h1>Medicinal Plant Identification</h1>
    <form action="/predict" method="post" enctype="multipart/form-data">
        <input type="file" name="file" accept="image/*">
        <input type="submit" value="Upload">
    </form>
    '''

# Define the prediction route
@app.route('/predict', methods=['POST'])
def predict():
    if 'file' not in request.files:
        print("No file part in request")
        return "No file uploaded"
    
    file = request.files['file']
    
    if file.filename == '':
        print("No file selected")
        return "No selected file"
    
    # Save the uploaded image to a temporary file
    filepath = os.path.join('uploads', file.filename)
    print(f"Attempting to save image at: {filepath}")  # Debugging path
    try:
        file.save(filepath)
        print(f"Image saved at: {filepath}")
    except Exception as e:
        print(f"Error saving the image: {e}")
        return "Error saving the image"

    # Load and preprocess the image
    try:
        img = image.load_img(filepath, target_size=(224, 224))  # Adjust to your input size
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0) / 255.0  # Normalize
        print("Image preprocessed")
    except Exception as e:
        print(f"Error in preprocessing the image: {e}")
        return "Error in preprocessing the image"

    # Make prediction
    try:
        predictions = model.predict(img_array)
        predicted_class_index = np.argmax(predictions)
        predicted_class_label = class_labels[predicted_class_index]
        print(f"Prediction made successfully: {predicted_class_label}")
    except Exception as e:
        print(f"Error in making prediction: {e}")
        return "Error in making prediction"

    # Return prediction result
    return f"Predicted Class: {predicted_class_label}"

# Run the Flask app using werkzeug
if _name_ == '_main_':
    run_simple('127.0.0.1', 5000, app)  # Use werkzeug to run the app
```
